```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.close("all")

df = pd.read_csv("house_price_Dống-Da_Hà-Nội_subdata.csv")
df
```

```python
df = df.dropna()
df.reset_index()
```

```python
df.describe()
```

```python
# Vẽ biểu đồ phân tích mối liên hệ giữa diện tích với giá nhà, 
# giữa số phòng ngủ với giá nhà, 
# giữa số toilet với giá nhà.
N = 14
sizes = np.random.rand(N)*200
colors = np.random.rand(N)*300

plt.scatter(df["price"], df["area"], s=sizes, c=colors)
plt.title("Relationship between housing area and price", fontsize = 16)
plt.xlabel("Price", fontsize = 14)
plt.ylabel("Area", fontsize = 14)
plt.show()
```

```python
# Nhìn chung, giá nhà tăng theo diện tích nhà. Một số trường hợp
# giá nhà thấp dù diện tích lớn, có thể khu vực xa dân cư,
# vị trí nhà không tốt (trong ngõ, hướng không tốt vv)
```

```python
# Vẽ biểu đồ phân tích mối liên hệ giữa diện tích với giá nhà, 
# giữa số phòng ngủ với giá nhà, 
# giữa số toilet với giá nhà.
plt.bar(df["bedroom"], df["price"], color = "green")
plt.title("Relationship between housing price and bedroom", fontsize = 16)
plt.xlabel("Number of bedroom", fontsize = 14)
plt.ylabel("Price", fontsize = 14)
plt.show()
```

```python
# Giá nhà ở khoảng trung bình thì thường có số lượng phòng ngủ từ 4 phòng ngủ trở lên.
# Giá nhà cao nhất khi có số phòng ngủ 2 phòng ngủ có thể thấy
# 2 phòng ngủ khá phù hợp cho gia đình có 3.
```

```python
# Vẽ biểu đồ phân tích mối liên hệ giữa diện tích với giá nhà, 
# giữa số phòng ngủ với giá nhà, 
# giữa số toilet với giá nhà.
plt.bar(df["toilet"], df["price"], color = "orange")
plt.title("Relationship between housing price and bedroom", fontsize = 16)
plt.xlabel("Number of toilet", fontsize = 14)
plt.ylabel("Price", fontsize = 14)
plt.show()
```

```python
# Giá nhà cao nhất khi có số phòng toilet la 2 phòng, cũng liên
# quan đến 2 phòng ngủ ở trên. 
```
