```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df=pd.read_csv("shopeep_koreantop_clothing_shop_data.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df1=df.groupby(["join_year"])["shopid"].sum()
df1 = pd.DataFrame(df1)
df1.reset_index(inplace=True)
df1
```

```python
#Vẽ biểu đồ so sánh số lượng shop gia nhập theo các năm.

plt.bar(df["join_year"],df["shopid"])
plt.title("Number of attended shops")
plt.xlabel("Years", fontsize = 14)
plt.ylabel("Number of shops", fontsize = 14)

plt.show()
```

```python
# So luong cac cua hang tham gia tang tuyen tinh theo nam. 
```

```python
# Vẽ biểu đồ thể hiện mối quan hệ giữa tỉ lệ phản hồi với số lượt khách hàng đánh giá tốt.
N = 746
sizes = np.random.rand(N)*200
colors = np.random.rand(N)*300
plt.scatter(df["response_rate"],df["rating_good"],s=sizes, c=colors)
plt.title("Relationship between response_rate and rating_good")
plt.xlabel("Response rate", fontsize = 14)
plt.ylabel("Rating good", fontsize = 14)
plt.yscale("log")
plt.show()
```

```python
# Ty le rating good  giam theo thoi gian phan hoi. Ve muc 50-50.
```

```python
# Vẽ biểu đồ thể hiện mối quan hệ giữa thời gian phản hồi 
# (đơn vị giây) với sốlượt khách hàng đánh giá xấu
df['Hour']=pd.to_datetime(df['response_time'],format=' %H:%M:%S').dt.hour
df['Minute']=pd.to_datetime(df['response_time'],format=' %H:%M:%S').dt.minute
df['Second']=pd.to_datetime(df['response_time'],format=' %H:%M:%S').dt.second
df["Total_second"] = df["Hour"]*3600 + df["Minute"]*60 + df["Second"]
```

```python
df
```

```python
# Vẽ biểu đồ thể hiện mối quan hệ giữa thời gian phản hồi (đơn vị giây) với số lượt khách hàng đánh giá xấu.
N = 746
sizes = np.random.rand(N)*200
colors = np.random.rand(N)*300
plt.scatter(df["Total_second"],df["rating_bad"],s=sizes, c=colors)
plt.title("Relationship between respond time (in second) and rating_bad")
plt.xlabel("Response time (in second)", fontsize = 14)
plt.ylabel("Rating bad", fontsize = 14)
plt.yscale("log")
plt.xscale("log")
plt.show()
```

```python
# ty le rating bad tang theo thoi gian phan hoi. Ve muc 50 - 50
```

```python
#Vẽ biểu đồ thể hiện xu hướng của số lượng shop gia nhập theo thời gian
df.info()
```

```python
df["join_day"]=df["join_day"].astype(str)
df["join_year"]=df["join_year"].astype(str)
df["Date"] = pd.to_datetime(df["join_day"] + " " + df["join_month"] +"," + " " + df["join_year"], format="%d %B, %Y")
```

```python
df.info()
```

```python
plt.scatter(df["Date"],df["shopid"])
plt.title("Number of attended shops as time")
plt.xlabel("Time", fontsize = 14)
plt.ylabel("Number of shops", fontsize = 14)

plt.show()
```

```python
# So luong shop gia nhap tang theo thoi gian
```

```python
# Vẽ biểu đồ thể hiện phân bố của điểm đánh giá trung bình.
plt.hist(df["rating_normal"],bins = 6)
plt.title("Distribution of rating_normal")
plt.xlabel("rating_nomal scale", fontsize = 14)
plt.ylabel("Number of rating_normal", fontsize = 14)
plt.yscale("log")
plt.show()

```
