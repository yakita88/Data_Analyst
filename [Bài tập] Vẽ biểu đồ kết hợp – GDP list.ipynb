```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df =pd.read_csv("GDPlist.csv", encoding = "ISO-8859-1")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df1 = df[(df["Country"]=="\xa0Vietnam") |
         (df["Country"]=="\xa0Indonesia") |
         (df["Country"]=="\xa0Cambodia") |
         (df["Country"]=="\xa0Thailand") |
         (df["Country"]=="\xa0Malaysia")]
```

```python
df1.reset_index()
```

```python
plt.style.use("ggplot")
labels = ["Cambodia","Indonesia","Malaysia","Thailand","Vietnam"]
explode = (0, 0, 0, 0, 0.5)
plt.pie(df1["GDP (millions of US$)"], labels = labels,
       autopct ="%1.5f%%", explode = explode)
plt.title("Vietnam's GDP in 5 countries", fontsize = 16)
plt.show()
```

```python
import matplotlib.pyplot as plt

x1 = df1["Country"]
y1 = df1["GDP (millions of US$)"]
labels = ["Cambodia","Indonesia","Malaysia","Thailand","Vietnam"]

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].bar(x1, y1, width = 0.5)
ax[0].set_yscale("log")
ax[0].tick_params(axis='y', size = 8)
ax[0].set_xticklabels(labels,rotation=90,fontname ='Arial')
ax[0].set_ylabel('GDP (millions of US$)', fontsize = 10, fontname ='Arial')

explode = (0, 0, 0, 0, 0.5)
ax[1].pie(df1["GDP (millions of US$)"],labels = labels, shadow = True,
        autopct ="%1.2f%%", explode = explode,textprops={'fontsize': 10, "fontname":"Arial"},
          wedgeprops={'linewidth': 1, 'linestyle': 'solid', 'edgecolor':'white'})

plt.suptitle("Vietnam's GDP in 5 countries", fontsize = 16)

plt.show()
```
