```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("FoodPrice_in_Turkey.csv", encoding = "ISO-8859-1")
df
```

```python
df1 = df[(df["Year"]==2019)&(df["Month"]==12)&(df["ProductName"] == "Milk (powder, infant formula) - Retail")].reset_index()
df2 = df[(df["Year"]==2019)&(df["Month"]==12)&(df["ProductName"]== "Fuel (gas) - Retail")].reset_index()

data1 = pd.DataFrame({"x":df1["Place"], "y1":df1["Price"], "y2": df2["Price"]})
data1
# df1 = df[(df['Year'] == 2019) & (df['Month'] == 12) & (df['ProductName'] == 'Milk (powder, infant formula) - Retail')].reset_index()
# df2 = df[(df['Year'] == 2019) & (df['Month'] == 12) & (df['ProductName'] == 'Fuel (gas) - Retail')].reset_index()
# data1 = pd.DataFrame({'x': df1['Place'], 'y1': df1['Price'], 'y2': df2['Price']})
```

```python
data1.plot(x = "x", y = ["y1", "y2"], kind = "bar")
plt.title("Milk and Gas Price in 12/2019", fontsize = 16, color = "r")
plt.xlabel("Place", fontsize = 14)
plt.ylabel("Price", fontsize = 14)
plt.show()
```

```python
df11 = df[(df['Place'] == 'National Average') & (df['Year'] == 2016) & (df['ProductName'] == 'Milk (powder, infant formula) - Retail')]
df12 = df[(df['Place'] == 'National Average') & (df['Year'] == 2016) & (df['ProductName'] == 'Fuel (gas) - Retail')]
df13 = df[(df['Place'] == 'National Average') & (df['Year'] == 2018) & (df['ProductName'] == 'Milk (powder, infant formula) - Retail')]
df14 = df[(df['Place'] == 'National Average') & (df['Year'] == 2018) & (df['ProductName'] == 'Fuel (gas) - Retail')]
df15 = df[(df['Place'] == 'National Average') & (df['Year'] == 2019) & (df['ProductName'] == 'Milk (powder, infant formula) - Retail')]
df16 = df[(df['Place'] == 'National Average') & (df['Year'] == 2019) & (df['ProductName'] == 'Fuel (gas) - Retail')]

df21 = df[(df['Place'] == 'National Average') & (df['ProductName'] == 'Fuel (gas) - Retail') & (df['Year'] == 2016)]
df22 = df[(df['Place'] == 'National Average') & (df['ProductName'] == 'Milk (powder, infant formula) - Retail') & (df['Year'] == 2016)]
df23 = df[(df['Place'] == 'National Average') & (df['ProductName'] == 'Fuel (gas) - Retail') & (df['Year'] == 2018)]
df24 = df[(df['Place'] == 'National Average') & (df['ProductName'] == 'Milk (powder, infant formula) - Retail') & (df['Year'] == 2018)]
df25 = df[(df['Place'] == 'National Average') & (df['ProductName'] == 'Fuel (gas) - Retail') & (df['Year'] == 2019)]
df26 = df[(df['Place'] == 'National Average') & (df['ProductName'] == 'Milk (powder, infant formula) - Retail') & (df['Year'] == 2019)]
```

```python
fig, ax = plt.subplots(3, 2)
ax[0][0].plot(df11['Month'], df11['Price'], marker = '*', label = 'Milk-2016')
ax[0][0].plot(df12['Month'], df12['Price'], marker = 's', label = 'Gas-2016')
ax[0][0].set_ylabel('Price')
ax[0][0].set_xticklabels([])
ax[0][0].set_title('2016')

ax[1][0].plot(df13['Month'], df13['Price'], marker = '*', label = 'Milk-2018')
ax[1][0].plot(df14['Month'], df14['Price'], marker = 's', label = 'Gas-2018')
ax[1][0].set_ylabel('Price')
ax[1][0].set_xticklabels([])
ax[1][0].set_title('2018')

ax[2][0].plot(df15['Month'], df15['Price'], marker = '*', label = 'Milk')
ax[2][0].plot(df16['Month'], df16['Price'], marker = 's', label = 'Gas')
ax[2][0].set_ylabel('Price')
ax[2][0].set_xlabel('Month')
ax[2][0].legend()
ax[2][0].set_title('2019')

ax[0][1].scatter(df21['Price'], df22['Price'])
ax[0][1].set_title('2016')
ax[0][1].set_ylabel('Rice')
ax[0][1].set_xticklabels([])

ax[1][1].scatter(df23['Month'], df24['Price'])
ax[1][1].set_title('2018')
ax[1][1].set_xticklabels([])

ax[2][1].scatter(df25['Month'], df26['Price'])
ax[2][1].set_title('2019')
ax[2][1].set_xlabel('Gas')
fig.tight_layout()
plt.show()
```
