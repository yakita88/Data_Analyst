```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("house_price_Dống-Da_Hà-Nội_subdata.csv", encoding = "ISO-8859-1")
df
```

```python
df.info()
```

```python
df["toilet"]=df["toilet"].fillna(method="ffill")
df["bedroom"]=df["bedroom"].fillna(method="ffill")
df["floor"]=df["floor"].fillna(method="ffill")
df["price"]=df["price"].fillna(method="ffill")
df["area"]=df["area"].fillna(method="ffill")
```

```python
df.info()
```

```python
df1 = df[["area","price","toilet","bedroom","floor"]]
df1
```

```python
fig, ax = plt.subplots(3, 1)
ax[0].scatter(df1['price'], df1['area'], marker = '*')
ax[0].set_ylabel('area')
ax[0].set_xticklabels(df1["price"])
ax[0].set_title('Relationship between housing area and price')

ax[1].scatter(df1['price'], df1['toilet'],marker = '.')
ax[1].set_ylabel('toilet')
ax[1].set_xticklabels(df1["price"])
ax[1].set_title('Relationship between housing price and number of toilet')

ax[2].scatter(df1['price'], df1['bedroom'],marker = 's')
ax[2].set_ylabel('bedroom')
ax[2].set_xticklabels(df1["price"])
ax[2].set_title('Relationship between housing price and number of bedroom')

# ax[0][1].scatter(df21['Price'], df22['Price'])
# ax[0][1].set_title('2016')
# ax[0][1].set_ylabel('Rice')
# ax[0][1].set_xticklabels([])

# ax[1][1].scatter(df23['Month'], df24['Price'])
# ax[1][1].set_title('2018')
# ax[1][1].set_xticklabels([])

# ax[2][1].scatter(df25['Month'], df26['Price'])
# ax[2][1].set_title('2019')
# ax[2][1].set_xlabel('Gas')
fig.tight_layout()
plt.show()
```
