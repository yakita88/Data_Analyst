```python
import pandas as pd
import numpy as np

df = pd.read_csv("OnlineRetail.csv", encoding ="ISO-8859-1")
df
```

```python
df.info()
# 8 columns and 541909 rows
# InvoiceNo, StockCode, Description, InvoiceDate <object> - Norminal 
# CustomerID <objiect> --> Norminal
# Country <float64> -- Norminal
# Quantity <int64>; UnitPrice <float64> -- Ratio
```

```python
df1 = df.pivot_table(values =["Quantity"], index=["InvoiceNo","Country"],aggfunc ="sum")
df1
```

```python
df2 = df.pivot_table(values =["Quantity"], index=["CustomerID"],aggfunc ={"max","min"})
df2
```

```python
df2 = df.pivot_table(values =["Quantity","UnitPrice"], index=["StockCode"],aggfunc ={"Quantity":"sum","UnitPrice":"mean"})
df2
```

```python
df3 = df.pivot_table(values =["Quantity"], index=["InvoiceDate"],aggfunc ="sum")
df3
```
