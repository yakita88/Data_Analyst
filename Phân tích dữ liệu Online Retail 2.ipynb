```python
import pandas as pd
import numpy as np

df = pd.read_csv("OnlineRetail.csv", encoding = "ISO-8859-1")
df
```

```python
df["InvoiceDate"].unique()
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"],format = "%m/%d/%Y %H:%M")
df["Month"]=pd.to_datetime(df["InvoiceDate"],format = "%m/%d/%Y %H:%M").dt.month
df
```

```python
df.info()
```

```python
conditions = [(df["Month"]>=1) & (df["Month"] <=3),
             (df["Month"] >= 4) & (df["Month"] <= 6),
             (df["Month"] >= 7) & (df["Month"] <= 9),
              (df["Month"] >= 10) & (df["Month"] <= 12)]
choices = [1,2,3,4]
df['Quarter'] = np.select(conditions, choices, default='1')
df
```

```python
df["Month"].unique()
```

```python
df["Amount"]=df["Quantity"]*df["UnitPrice"]
df
```

```python
conditions = [(df["Country"]=="United Kingdom")&(df["Quarter"]==4),
             (df["Country"] == "France"),
             (df["Country"]!="United Kingdom")&(df["Country"] != "France")]
choices = [0.1,0.05,0]
df["Discount"] = np.select(conditions,choices,default=0.1)
df
```

```python
df["Total"]=df["Amount"]-df["Discount"]*df["Amount"]
df
```
